##### Making changes to the data

```python
# Creating a new column as total of the values in Attack and defence columns

pokemon_df['Points'] = pokemon_df.loc[:,'HP':'Speed'].sum(axis = 1)
```

```python
pokemon_df['Best_skill'] = pokemon_df.loc[:,'HP':'Speed'].idxmax(axis = 1)
```

```python
pokemon_df.head()
```

#### Filtering of data

```python
#Filter where type 1 is grass

pokemon_df[pokemon_df['Type 1'] == 'Grass']
```

```python
##### Filtering based on two or logical operations

# Filter the dataframe where Attack > 60 and Defense > 80 or speed > 100

pokemon_df[((pokemon_df['Attack'] > 60) & (pokemon_df['Defense'] > 80))  | (pokemon_df['Speed'] > 100)]
```

```python
#### Filerting with str (can also pass complicated equations, regex and more)

pokemon_df[pokemon_df.Name.str.contains(r'[0-9]+')]
```

```python
#### Str operation can also be useful to replace data in the column

pokemon_df.Best_skill = pokemon_df.Best_skill.map(lambda x : x.replace('Sp.','Special'))
```

```python
#### Find Type 1 being Grass or Fire

pokemon_df[pokemon_df['Type 1'].str.contains('Grass|Fire', regex = True)]
```

### Aggregate Statistics

```python
#Group by

pokemon_df.groupby('Pok_index')['Name'].count()
```

```python
pokemon_df.groupby('Best_skill')['Points'].mean()
```

```python
pokemon_df.groupby('Type 1')['Type 1'].count().sort_values(ascending  = False)
```
